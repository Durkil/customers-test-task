```python
import pandas as pd
```

```python
df = pd.read_csv('Customer.csv')
```

```python
df.head(10)
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date']).dt.strftime('%Y-%m-%d')
```

```python
df.head(30)
```

```python
df = df[df['email'].str.contains(r'^[\w\.-]+@[\w\.-]+\.com$', regex=True)]
```

```python
df.head(30)
```

```python
df['domain'] = df['email'].str.split('@').str[1]
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset='email', keep='first')
```

```python
df.to_csv('New_Customer.csv', index=False)
```
